```python
import torch
import numpy as np
import matplotlib.pyplot as plt

from blackboxopt import Objective, EvaluationSpecification
from benchmarks_multitask import Forrester
from optimizers.banner import BaNNER_WS

from plotting.banner_plots import plot_meta_mean_var, plot_metadata
```

```python
benchmark = Forrester(n_data_per_task=[32] * 128)
meta_data = benchmark.get_meta_data(distribution='sobol')
```

## Meta-data

```python
fig = plt.figure(figsize=(12, 4))

ax_meta = plt.subplot2grid((1, 2), (0, 0))
plot_metadata(benchmark, meta_data, ax_meta)

ax_mean_var = plt.subplot2grid((1, 2), (0, 1))
plot_meta_mean_var(benchmark, meta_data, ax_mean_var)
```

## Meta-learning

```python
meta_model_config = {
    "num_layers": 5,
    "num_features": 2,
    "num_hidden_units": 64,
    "device": "cpu"
}
train_config = {
    "num_epochs": 100,
    "batch_size": 64,
}

model = BaNNER_WS(benchmark.search_space, **meta_model_config)
model.meta_fit(meta_data=meta_data, **train_config)
```

```python
x_config = np.linspace(-1, 1, 100).reshape(-1, 1)
x_numerical = []
for x in x_config:
    x_numerical.append(benchmark.search_space.to_numerical({'x': x}))
x_numerical = np.array(x_numerical)
```

```python
mean, var = model.predict(x_numerical)

with torch.no_grad():
    embeddings = model.meta_blr.meta_model.task_embedding(torch.LongTensor(list(range(1, 129))))
```

```python
fig = plt.figure(figsize=(12, 4))
ax_meta = plt.subplot2grid((1, 2), (0, 0))
ax_meta.plot(x_config, mean)
ax_meta.fill_between(
    x_config.squeeze(),
    (mean + 2*np.sqrt(var)).squeeze(),
    (mean - 2*np.sqrt(var)).squeeze(),
    alpha=.3
)
ax_meta.set_xlim(*benchmark.search_space.get_continuous_bounds())

ax_embedding = plt.subplot2grid((1, 2), (0, 1))
ax_embedding.scatter(embeddings[:, 0], embeddings[:, 1], label="embedding samples")

prior_samples = torch.randn(model.meta_blr.meta_model.task_embedding.weight[1:].shape)
ax_embedding.scatter(prior_samples[:, 0], prior_samples[:, 1], label="prior samples")
ax_embedding.legend();
```

```python
for i in range(20, 23):
    embedding_train = model.meta_blr.meta_model.task_embedding(torch.tensor(i)).detach()
    mean, var = model.predict(x_numerical, embedding=embedding_train)

    fig = plt.figure(figsize=(12, 4))
    ax_meta = plt.subplot2grid((1, 2), (0, 0))
    ax_meta.plot(x_config, mean)
    ax_meta.fill_between(
        x_config.squeeze(),
        (mean + 2*np.sqrt(var)).squeeze(),
        (mean - 2*np.sqrt(var)).squeeze(),
        alpha=.3
    )
    ax_meta.set_xlim(*benchmark.search_space['x']['parameter'].bounds);
```
